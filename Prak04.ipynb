```python
s1 = ["<s> franz ist kaiser <e>"]
s2 = ["<s> franz mag fussball <e>"]
s3 = ["<s> kaiser isst semmeln <e>"]
s4 = ["sissi"]

corpus = s1 + s2 + s3 + s4
vocab = {"franz", "ist", "kaiser", "mag", "fussball", "isst", "semmeln"}
```

```python
from collections import defaultdict
def estimate_unigram(data):
    counts = defaultdict(int)
    for example in data:
        for w in example:
            if w not in ["<s>","<e>"]:
                counts[w] += 1
    return counts
```

```python

model1 = estimate_unigram(corpus)
ss1 = ["<s> franz isst semmeln <e>"]
ss2 = ["<s> franz ist semmeln <e>"]
```

```python
import math 
import numpy as np
def unigram_sentence_logp(model, sentence):
    ttc = sum(model.values())
    sum_of_log2 = 0 
    for word in sentence:
        if word not in model: #words that are not in the model will be skipped later in unigram_perplexity 
           return -np.inf      # will be unecessary when smoothing is implemented 

        p_of_w = model[word] / ttc
        log2_of_w = math.log2(p_of_w)
        sum_of_log2 += log2_of_w
    return sum_of_log2
```

```python
print("logprobability of ss1:", unigram_sentence_logp(model1, ss1))
print("logprobability of ss2:", unigram_sentence_logp(model1, ss2))
#model returns this because it may have not seen certain words in corpus
#possible that I haven't skipped over s and e correctly
```

```python
import math 
import numpy as np
def unigram_perplexity(model, test_data):
    total_log = 0
    total_words = 0
    for example in test_data:
        words = example
        logp = unigram_sentence_logp(model, words)
        if logp == np.inf or logp == -np.inf:
            continue

        total_log += logp
        total_words += len(words)

    avg_logp = total_log / total_words
    perplexity = 2**(-avg_logp)
    return perplexity 
```

```python
print("perplexity of ss1:", unigram_perplexity(model1, ss1))
print("perplexity of ss2:", unigram_perplexity(model1, ss2))
#To get more reasonable results maybe introduce smoothing
```
